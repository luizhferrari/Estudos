```python

import numpy as np
import pandas as pd
import string
import warnings
warnings.filterwarnings("ignore")
```

```python
import requests

#A url que você quer acesssar
url = 'https://www.fundamentus.com.br/resultado.php'

#Informações para fingir ser um navegador
header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}
#juntamos tudo com a requests
r = requests.get(url, headers=header)
#E finalmente usamos a função read_html do pandas
#Ajustando os decimais e milhares
df = pd.read_html(r.text, decimal=',', thousands='.')[0]
```

```python
df
```

```python
#Ajustando as colunas que possuem porcentagens
for coluna in ['Div.Yield', 'Mrg Ebit', 'Mrg. Líq.', 'ROIC', 'ROE', 'Cresc. Rec.5a']:
    df[coluna] = df[coluna].str.replace('.', '')
    df[coluna] = df[coluna].str.replace(',', '.')
    df[coluna] = df[coluna].str.rstrip('%').astype('float')/100
```

```python
df
```

```python
#Ranking de P/L
PL = pd.DataFrame()
PL['Papel'] = df.sort_values(by= ['P/L'], ascending=True)['Papel'].values
PL['P/L'] = df.sort_values(by= ['P/L'], ascending=True)['P/L'].values
PL.head()
```

```python
#filtrando acoes que possuem liquidez media de 1000000 por 2 meses
df = df[df['Liq.2meses'] > 2000000]
```

```python
df.shape
```

```python
#Formula Mágica inventada por alguém
# A ideia é um Ranking de EV/EBIT e ROIC e depois somar as posiçoes de cada uma das acoes
# O menor valor são as empresas que deve comprar
ranking = pd.DataFrame()
ranking['pos'] = range(1,151)
ranking['EV/EBIT'] = df[df['EV/EBIT']>0].sort_values(by= ['EV/EBIT'])['Papel'][:150].values
ranking['ROIC'] = df.sort_values(by= ['ROIC'], ascending=False)['Papel'][:150].values

#Pegando as posiçoes no ranking de cada ação
a = ranking.pivot_table(columns= 'EV/EBIT', values='pos')
b = ranking.pivot_table(columns= 'ROIC', values='pos')
t = pd.concat([a,b])

#Somando as posiçoes do EV/EBIT e ROIC
rank = t.dropna(axis=1).sum()

#Pegando apenas as 15 primeiras
rank.sort_values()[:15]
```
